```python
import pandas as pd
import numpy as np
```

```python
backward_overlap, no_overlap, forward_overlap = [], [], []
```

```python
f = open("./output.txt", "r")
```

```python
while True:
    line = f.readline()
    if not line:
        break

    tokens = line.split()

    backward_overlap.append(float(tokens[0]))
    no_overlap.append(float(tokens[1]))
    forward_overlap.append(float(tokens[2]))
f.close()
```

```python
b_ov = np.array(backward_overlap)
n_ov = np.array(no_overlap)
f_ov = np.array(forward_overlap)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(16,8), facecolor='white')
ax = fig.add_subplot(111)

plt.tight_layout()

plt.xlabel('Compute size scale', fontsize=20)
plt.ylabel('Single iteration time', fontsize=20)


ax.plot(b_ov, label='backward overlap', color='blue')
ax.plot(f_ov, label='forward overlap', color='red')
ax.plot(n_ov, label='no overlap', color='green')

plt.legend()
plt.show()
```
